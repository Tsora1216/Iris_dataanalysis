```python
import seaborn as sns
# 
g = sns.pairplot(df, hue = 'species')
g._legend.remove()
g.fig.legend(
    handles = g._legend_data.values(),
    labels = g._legend_data.keys(),
    loc = 'lower center', ncol = 3,
    title = 'species')
g.fig.subplots_adjust(bottom = 0.12)
```
